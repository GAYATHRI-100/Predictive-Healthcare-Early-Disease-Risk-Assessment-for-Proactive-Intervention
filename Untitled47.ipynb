```python

import pandas as pd
import json
import re
import io
from google.colab import files

print("Please upload users.json, orders.csv, and restaurants.sql")
uploaded = files.upload()

users_df = pd.read_json(io.BytesIO(uploaded['users.json']))


orders_df = pd.read_csv(io.BytesIO(uploaded['orders.csv']))
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'], format='%d-%m-%Y')

sql_content = uploaded['restaurants.sql'].decode('utf-8')
pattern = r"INSERT INTO restaurants VALUES \((\d+), '(.*?)', '(.*?)', (.*?)\);"
matches = re.findall(pattern, sql_content)
restaurants_df = pd.DataFrame(matches, columns=['restaurant_id', 'restaurant_name', 'cuisine', 'rating'])
restaurants_df['restaurant_id'] = restaurants_df['restaurant_id'].astype(int)
restaurants_df['rating'] = restaurants_df['rating'].astype(float)

merged_df = orders_df.merge(users_df, on='user_id', how='left')
merged_df = merged_df.merge(restaurants_df[['restaurant_id', 'cuisine', 'rating']], on='restaurant_id', how='left')

print("\n--- Analysis Results ---\n")


avg_order_cuisine = merged_df.groupby('cuisine')['total_amount'].mean()
print(f"1. Cuisine with highest Avg Order Value: {avg_order_cuisine.idxmax()} (₹{avg_order_cuisine.max():.2f})")

user_totals = merged_df.groupby('user_id')['total_amount'].sum()
users_gt_1000 = (user_totals > 1000).sum()
print(f"2. Distinct users with total orders > ₹1000: {users_gt_1000}")


bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0 – 3.5', '3.6 – 4.0', '4.1 – 4.5', '4.6 – 5.0']
merged_df['rating_range'] = pd.cut(merged_df['rating'], bins=bins, labels=labels, include_lowest=True)
revenue_by_rating = merged_df.groupby('rating_range')['total_amount'].sum()
print(f"3. Rating range with highest revenue: {revenue_by_rating.idxmax()}")

gold_orders = merged_df[merged_df['membership'] == 'Gold']
avg_gold_city = gold_orders.groupby('city')['total_amount'].mean()
print(f"4. Top Gold member city (Avg Order Value): {avg_gold_city.idxmax()}")


gold_pct = (len(gold_orders) / len(orders_df)) * 100
print(f"5. Percentage of orders by Gold members: {round(gold_pct)}%")


revenue_comb = merged_df.groupby(['membership', 'cuisine'])['total_amount'].sum()
print(f"6. Highest Revenue Combination: {revenue_comb.idxmax()}")


merged_df['quarter'] = merged_df['order_date'].dt.quarter
revenue_quarter = merged_df.groupby('quarter')['total_amount'].sum()
print(f"7. Quarter with highest revenue: Q{revenue_quarter.idxmax()}")
```
